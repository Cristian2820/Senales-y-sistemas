##Cero y pendiente para acondicionamiento de señales

Simule un proceso de discretización con $F_s \geq 2 F_o.$ Dibuje el resultado de la señal discretizada utilizando la función stem.

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear vector tiempo por rango y por número de muestras
K = 200  # número de muestras
ti = -15  # tiempo inicial referencia [s]
tf = 10  # tiempo final [s]

# Vector de tiempo equiespaciado
tv = np.linspace(start=ti, stop=tf, num=K)
print(tv)
print(tv.shape)

# Tiempo de muestreo se define según el rango de trabajo en segundos y la cantidad de muestras
Ts = (tf - ti) / K
print(Ts)

# Parámetros parábola
a = 4
b = -3
xv = a * (tv**2) + b  # Creación de parábola en tiempo discreto
print(xv)
print(xv.shape)

# Gráfica de la parábola
plt.plot(tv, xv, 'r.')  # Vector en eje horizontal, vector eje vertical, color y marcador
plt.xlabel('Tiempo [s]')
plt.ylabel('$x[t_k]$')  # $ $ ecuación LaTeX
plt.title('Señal Continua')
plt.show()  # Mostrar gráfica

# Definir la frecuencia de muestreo
F_o = 1  # Frecuencia de la señal original (asumida)
F_s = 2 * F_o  # Frecuencia de muestreo (Nyquist)
Ts_new = 1 / F_s  # Nuevo tiempo de muestreo

# Crear un nuevo vector de tiempo para la señal discretizada
tv_new = np.arange(ti, tf, Ts_new)
xv_new = a * (tv_new**2) + b  # Creación de parábola en tiempo discreto con el nuevo tiempo de muestreo

# Gráfica de la señal discretizada
plt.stem(tv_new, xv_new, use_line_collection=True)
plt.xlabel('Tiempo [s]')
plt.ylabel('$x[t_k]$')  # $ $ ecuación LaTeX
plt.title('Señal Discretizada')
plt.show()  # Mostrar gráfica
```

# Función Seno

Simule un proceso de discretización con $F_s \geq 2 F_o.$ Dibuje el resultado de la señal discretizada utilizando la función stem.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la señal continua
K = 500  # número de muestras
T_o = 1/60  # periodo de la función sin
A = 120  # amplitud
ti = 0  # tiempo inicial
tf = T_o  # tiempo final

# Vector de tiempo equiespaciado
tv = np.linspace(ti, tf, num=K)  # vector de datos en tiempo
xv = A * np.sin(2 * np.pi * tv / T_o)  # vector de amplitud

# Gráfica de la señal continua
plt.plot(tv, xv)
plt.xlabel('Tiempo [s]')
plt.ylabel('$A\\sin[\\omega t_k]$')
plt.title('Señal Continua')
plt.show()

# Definir la frecuencia de muestreo
F_o = 1 / T_o  # Frecuencia de la señal original
F_s = 10 * F_o  # Frecuencia de muestreo (al menos 10 veces F_o para mejor visualización)
Ts_new = 1 / F_s  # Nuevo tiempo de muestreo

# Crear un nuevo vector de tiempo para la señal discretizada
tv_new = np.arange(ti, tf, Ts_new)
xv_new = A * np.sin(2 * np.pi * tv_new / T_o)  # Creación de la señal muestreada

# Gráfica de la señal discretizada
plt.stem(tv_new, xv_new, use_line_collection=True)
plt.xlabel('Tiempo [s]')
plt.ylabel('$A\\sin[\\omega t_k]$')
plt.title('Señal Discretizada')
plt.show()
```

Consultar el funcionamiento de los métodos cdist y argmin.

*   Cdist: La función cdist calcula las distancias entre cada par de puntos de dos conjuntos de puntos. Es parte del módulo scipy.spatial.distance y es muy útil para diversas aplicaciones de análisis de datos, como clustering, machine learning, y más.

*   Argmin: La función argmin devuelve los índices del valor mínimo a lo largo de un eje. Es parte del módulo numpy y es muy útil para encontrar rápidamente el índice del valor más pequeño en un array.


Realice una función mediante el ambiente def que permita realizar los procesos de discretización y cuantización (usando cdist y argmin) en cascada para un conversor análogo-digital de 3 bits con entrada análoga de 0 a 10 [v]. Presente las gráficas de la señal original, señal con cero y pendiente, señal discretizada, y señal digital.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist

def ad_converter(signal, time, num_bits=3, v_min=0, v_max=10, fs_factor=10):
    """
    Realiza el proceso de discretización y cuantización en cascada para un conversor análogo-digital.

    Parámetros:
    - signal: Señal continua de entrada.
    - time: Vector de tiempo para la señal de entrada.
    - num_bits: Número de bits del conversor (por defecto 3).
    - v_min: Voltaje mínimo de entrada (por defecto 0).
    - v_max: Voltaje máximo de entrada (por defecto 10).
    - fs_factor: Factor para determinar la frecuencia de muestreo (por defecto 10 veces la frecuencia de la señal).

    Retorna:
    - time_samples: Vector de tiempo de la señal discretizada.
    - signal_discretized: Señal discretizada.
    - signal_quantized: Señal cuantizada.
    """
    # Parámetros de cuantización
    num_levels = 2 ** num_bits
    quant_step = (v_max - v_min) / num_levels
    quant_levels = np.arange(v_min + quant_step/2, v_max, quant_step)

    # Discretización
    T_o = time[-1] - time[0]
    F_o = 1 / T_o
    F_s = fs_factor * F_o
    Ts_new = 1 / F_s
    time_samples = np.arange(time[0], time[-1], Ts_new)
    signal_discretized = np.interp(time_samples, time, signal)

    # Cuantización
    distances = cdist(signal_discretized[:, np.newaxis], quant_levels[:, np.newaxis])
    quant_indices = np.argmin(distances, axis=1)
    signal_quantized = quant_levels[quant_indices]

    return time_samples, signal_discretized, signal_quantized

# Señal de ejemplo
K = 500  # número de muestras
T_o = 1  # periodo de la señal
A = 5  # amplitud de la señal
ti = 0  # tiempo inicial
tf = T_o  # tiempo final
tv = np.linspace(ti, tf, num=K)  # vector de tiempo
xv = A * np.sin(2 * np.pi * tv / T_o) + 5  # señal continua (desfasada para que esté en el rango 0-10)

# Aplicar el conversor análogo-digital
time_samples, signal_discretized, signal_quantized = ad_converter(xv, tv, num_bits=3, v_min=0, v_max=10, fs_factor=10)

# Gráficas
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(tv, xv)
plt.title('Señal Original')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [V]')

plt.subplot(2, 2, 2)
plt.plot(tv, xv - 5)
plt.title('Señal con Cero y Pendiente')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [V]')

plt.subplot(2, 2, 3)
plt.plot(tv, xv, label='Original')
plt.stem(time_samples, signal_discretized, 'r', markerfmt='ro', basefmt=" ", use_line_collection=True, label='Discretizada')
plt.title('Señal Discretizada')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [V]')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(tv, xv, label='Original')
plt.stem(time_samples, signal_quantized, 'g', markerfmt='go', basefmt=" ", use_line_collection=True, label='Digital')
plt.title('Señal Digital')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [V]')
plt.legend()

plt.tight_layout()
plt.show()
```
