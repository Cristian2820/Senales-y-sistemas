```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la seal continua x(t)
def x(t):
    return 3 * np.cos(1000*np.pi*t) + 5 * np.sin(2000*np.pi*t) + 10 * np.cos(11000*np.pi*t)

# Frecuencia de muestreo
fs = 5000  # 5 kHz

# Duracin de la seal continua
duracin = 0.02  # 20 ms

# Crear el eje de tiempo para la seal continua
t_continuo = np.arange(0, duracin, 1/fs)

# Muestrear la seal continua
x_discreto = x(t_continuo)

# Crear el eje de tiempo para la seal discreta
t_discreto = np.arange(0, duracin, 1/fs)

# Graficar la seal continua y la seal discreta
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(t_continuo, x(t_continuo))
plt.title('Seal Continua x(t)')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.stem(t_discreto, x_discreto)
plt.title('Seal Discreta x[n]')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.grid(True)

plt.tight_layout()
plt.show()
```
